# Step 1: Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("insight.csv")
```

```python
# Create readable month labels
df["month_label"] = df["order_year"].astype(str) + "-" + df["order_month"].astype(str).str.zfill(2)
```

```python
df.head()
```

# Line chart for revenue trend

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="month_label", y="total_revenue", marker="o", color="green")
plt.title("Monthly Total Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Bar chart for order volume trend

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x="month_label", y="total_orders", color="skyblue")
plt.title("Monthly Order Volume Trend")
plt.xlabel("Month")
plt.ylabel("Total Orders")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
